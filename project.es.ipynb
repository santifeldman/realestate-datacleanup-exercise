#### Ejercicio 15. Traza un gráfico de barra de la media de los precios y explica lo que observas (debes usar el subconjunto obtenido de la pregunta 14) (★★★)

Imprima un gráfico de barra de la media de precios y escriba en la celda Markdown un breve análisis del sector.

```python
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('assets/real_estate.csv', sep=';')
df = pd.DataFrame(ds)
south_belt = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']

filtered_df = df[df['level5'].isin(south_belt)]
mean_prices = filtered_df.groupby('level5')['price'].mean().sort_values()

plt.figure(figsize=(10, 6))
mean_prices.plot(kind='bar', color='orange')
plt.title('Precios')
plt.xlabel('Población')
plt.ylabel('Precio Promedio(€)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

En este gráfico de barras podemos observar el valor promedio de la vivienda en las distintas poblaciones que componen el cinturón sur de Madrid. 
Nótese que el valor promedio en Fuenlabrada es de 180000 Euros, mientras que Getafe es la población con el precio medio más elevado, en torno a 250000 Euros.

#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in the question 14 (★★☆)

Hint: Create a new column called `pps` (price per square) and then analyse the values

```python
ds = pd.read_csv('assets/real_estate.csv', sep=';')
df = pd.DataFrame(ds)

south_belt = ['Getafe', 'Alcorcón']
filtered_df = df[df['level5'].isin(south_belt)].copy()

filtered_df.loc[:,'pps'] = filtered_df['price'] / filtered_df['surface']

getafe_pps = filtered_df[filtered_df['level5']=='Getafe']
alcorcon_pps = filtered_df[filtered_df['level5'] =='Alcorcón']

getafe_pps_mean = getafe_pps['pps'].mean()
alcorcon_pps_mean = alcorcon_pps['pps'].mean()
print (f'El precio promedio por metro cuadrado en Getafe es de {getafe_pps_mean} euros. En Alcorcón el precio promedio por metro cuadrado es de {alcorcon_pps_mean} euros. Como podemos observar el precio/m2 en Alcorcón es ligeramente más elevado que en Getafe')
```
